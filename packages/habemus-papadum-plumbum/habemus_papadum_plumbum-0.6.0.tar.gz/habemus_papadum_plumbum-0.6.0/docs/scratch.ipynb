```python
from pdum.plumbum import pb, apb
```

```python
@pb
def add1(x):
    return x + 1


@apb
async def aadd1(x):
    return x + 1
```

```python
3 > add1 | add1 | add1
```

```python
3 > add1 | add1 | add1 | aadd1 | aadd1 | aadd1
```

```python
await (3 > add1 | add1 | add1 | aadd1 | aadd1 | aadd1)
```
