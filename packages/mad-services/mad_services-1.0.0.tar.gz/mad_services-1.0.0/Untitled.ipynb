```python
import io
import MDAnalysis as mda
from MDAnalysis.tests.datafiles import PDB_small

# Load a simple structure
u = mda.Universe(PDB_small)

# Select atoms
ag = u.atoms

# Create a BytesIO buffer
buf = io.BytesIO()

# Write to the buffer (must specify format explicitly)
ag.write(buf, format="PDB")

# Get the data from memory
pdb_bytes = buf.getvalue()
print(f"Wrote {len(pdb_bytes)} bytes")
print(pdb_bytes.decode("utf-8")[:200])  # show first 200 chars
```
