```python
# Core imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Statistical tools
from scipy import stats
from statsmodels.tsa.stattools import adfuller
from sklearn.cluster import KMeans

# Hidden Regime imports
from hidden_regime.data import FinancialDataLoader
from hidden_regime.models import HiddenMarkovModel
from hidden_regime.config import HMMConfig, FinancialDataConfig

# Plotting style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")
print(f"Notebook execution time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
```

```python
CURRENT_DATE = datetime.now()
```

```python
def analyze(ticker : str, end_date : datetime, num_days : int) -> None:
    start_date = end_date - timedelta(days=num_days)    
    fdl_config = FinancialDataConfig(ticker=ticker, end_date=end_date, num_samples=num_days)
    fdl = FinancialDataLoader(fdl_config)
    df = fdl.load_data()
    print(df.tail())
    print(df.index[0:5])
    print(f"Analyze {ticker}: from {df.index[0].strftime('%Y-%m-%d')} to {df.index[-1].strftime('%Y-%m-%d')} ({num_days} days)")

    def make_plot(name):
        f = plt.figure()
        df[name].plot.line()
        return f

    make_plot('price')
    make_plot('pct_change')
    make_plot('log_return')
```

```python
analyze('TSLA', CURRENT_DATE, 200)
```
